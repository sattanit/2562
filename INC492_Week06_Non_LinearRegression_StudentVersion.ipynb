```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
```

```python
rng = np.random.RandomState(0)

n_sample = 100
data_max, data_min = 1.4, -1.4
len_data = (data_max - data_min)

# sort the data to make plotting easier later
data = np.sort(rng.rand(n_sample) * len_data - len_data / 2)
noise = rng.randn(n_sample) * .3
target = data ** 3 - 0.5 * data ** 2 + noise
```

```python
# Enter your code here
full_data = pd.DataFrame({"input_feature" : data, "target" : target})
```

# Non-Linear regression

```python
# Create model
# Enter your code here
polynomial_regression = make_pipeline(PolynomialFeatures(degree=3, inculde_bias=False), LinearRegression())
# Train model
# Enter your code here
polynomial_regression.fit(data, target)
# Test model 
# Enter your code here
target_predicted = polynomial_regression.predict(data)
```

```python
# Error calculation
mse = mean_squared_error(target, target_predicted)
```

# Support Vector Regression


In Linear Regression we try to fit a line that minimizes the error as much as possible. But through Support Vector Machine well fit a line (for 2-d) or a hyperplane that tries to limit the error to a certain extent.

In SVR, the best fit line is the hyperplane that has the maximum number of points.

Screenshot 2023-03-03 070559.jpg

```python
# Create model (kernel = linear)
# Enter your code here
svr = SVR(kernel="linear")

# Train model
# Enter your code here
svr.fit(data, target)
# Test model 
# Enter your code here
target_predicted = svr.predict(data)
```

```python
# Error calculation
mse = mean_squared_error(target, target_predicted)
```

```python
ax = sns.scatterplot(data=full_data, x="input_feature", y="target", color="green", alpha=0.5)
ax.plot(data, target_predicted)
_ = ax.set_title("SVR (linear) - Mean squared error = {mse:.2f}")
```
